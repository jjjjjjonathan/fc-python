# Scatter Plots & Crosshairs in Matplotlib
---

We will take goals for and against from a season to create a scatter plot with Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
table = pd.read_csv('../csv/league_tables.csv')
table.head()
```

Matplotlib's `.plot()` is easy to use. Needs the data to plot along each of the axes and the plot type. In this case, the plot type is `'o'` to show that we want to plot markers.

```python
plt.plot(table['GF'], table['GA'], 'o')
```

Rather than directly plot the chart, we can create our chart area with the first line below, set its size, then add our features from there.

```python
# Create plot area
fig, ax = plt.subplots()

# Set plot size
fig.set_size_inches(7,5)

# plot chart as above, but change the plot type from 'o' to '*' - giving us stars
plt.plot(table['GF'], table['GA'], '*')

# add labels to chart area
ax.set_title('Goals for and against')
ax.set_xlabel('Goals For')
ax.set_ylabel('Goals Against')

plt.show()
```

Now, let's add a crosshair to our chart to display the average line. This should help our viewers see if a point is performing well or not.

We do this using `plt.plot()` again. We give it three arguments:

- `type`: `'k-'` gives us two instrusctions to plot with. K means black, - means draw a line
- `Start/End x locations` - give these two coordinates in a list. In this example, we calculate the average to get the coordinate.
- `start/End y locations` - again in a list. Below, they are `[90, 20]`

```python
fig, ax = plt.subplots()
fig.set_size_inches(7,5)

plt.plot(table['GF'], table['GA'], 'o')
plt.plot([table['GF'].mean(), table['GF'].mean()], [90,20], 'k:', lw=1)
plt.plot([20,90], [table['GA'].mean(), table['GA'].mean()], 'k:', lw=1)

ax.set_title('Goals for and against')
ax.set_xlabel('Goals For')
ax.set_ylabel('Goals Against')

plt.show()
```

We may want to classify the quadrants with text to label them.

Rather than `.set_title()`, we use `.text()`. You must give arguments for the x and y location, in addition to the text that you want to write. You could also give colour and size.

```python
fig, ax = plt.subplots()
fig.set_size_inches(7,5)

plt.plot(table['GF'], table['GA'], 'o')
plt.plot([table['GF'].mean(), table['GF'].mean()], [90,20], 'k:', lw=1)
plt.plot([20,90], [table['GA'].mean(), table['GA'].mean()], 'k:', lw=1)

ax.set_title('Goals for and against')
ax.set_xlabel('Goals For')
ax.set_ylabel('Goals Against')

ax.text(18,90, 'Poor attack, poor defence', color='red', size='8')
ax.text(67,20, 'Strong attack, strong defence', color='red', size='8')

plt.show()
```
