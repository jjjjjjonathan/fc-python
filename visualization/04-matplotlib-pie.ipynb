# Creating Pie Charts in Matplotlib
---

Pie charts should generall be avoided in most cases, unless your goal is to mislead or you have a strict use case for them.

That being said, just because we won't use much of something doesn't mean we don't need to know how it's done.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
league_wins = {
  'Team': ['Manchester United', 'Manchester City', 'Chelsea', 'Arsenal', 'Blackburn Rovers', 'Leicester City', 'Liverpool'],
  'Championships': [13, 6, 5, 3, 1, 1, 1]
}

df = pd.DataFrame(league_wins, columns=['Team', 'Championships'])
df
```

`plt.pie()` will plot the numbers in the championships column.

```python
plt.pie(df['Championships'])

# this next line just makes it so the plot looks a little cleaner in the notebook
plt.tight_layout()
```

A pie chart doesn't tell us a lot without labels

Matplotlib has good customization options, so let's add a title, labels, and colours into the arguments.

```python
# Create a list of colours used for the teams, in order

team_colours = ['#D9020D', '#84BBFF', '#004793', '#EF0107', '#009EE0', '#0053A0', '#D10022']

plt.pie(
  df['Championships'],
  # data labes are team names in dataFrame
  labels=df['Team'],
  # assign colours list
  colors=team_colours,
  # give a tidier angle to ur first data angle
  startangle=90
)

# add a title
plt.title('Premier League Titles')
plt.tight_layout()
```
