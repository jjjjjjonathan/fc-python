# Lollipop Charts in Matplotlib
---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

table = pd.read_csv('../csv/league_tables.csv')

# team colours in alphabetical order
team_colours = ['#EF0107', '#670E36', '#FF0000', '#0054A6', '#660808', '#004793', '#1B458F', '#0052CC', '#FAF7F7', '#0053A0', '#D10022', '#84BBFF', '#D9020D', '#000000', '#009900', '#FF0033', '#001C58', '#FFF71C', '#660033', '#FDB913']

table['colour'] = team_colours

table = table.sort_values(by=['Pts'], ascending=False)
table.head()
```

The `.hlines()` argument plots our data in horizontal lines. At its simplest, it needs three arguments, `y`, `xmin`, and `xmax`.

- `y`: the y coordinate for each bar. For a bar chart, we will most often use evenly spaced bars, so we provide a a sequence from 1-20 for a 20 lollipop chart. `np.arange` can provide this easily.
- `xmin`: the minimum value a bar can have. We'll set this to `0`.
- `xmax`: the maximum value a bar can have. We'll set this to the team that scored the most points

```python
plt.hlines(y=np.arange(1,21), xmin=0, xmax=table['Pts'])
```

```python
plt.hlines(y=np.arange(1,21), xmin=0, xmax=table['Pts'], color='skyblue') 
plt.plot(table['Pts'], np.arange(1,21), 'o')
plt.yticks(np.arange(1,21), table['Team'])
plt.show()
```

```python
plt.hlines(y=np.arange(1,21), xmin=0, xmax=table['Pts'],color=table['colour'])
plt.plot(table['Pts'], np.arange(1,21), 'o')
plt.yticks(np.arange(1,21), table['Team'])

plt.ylabel('Team')
plt.xlabel('Points')

plt.title('Premier League 21/22')
plt.show()
```
